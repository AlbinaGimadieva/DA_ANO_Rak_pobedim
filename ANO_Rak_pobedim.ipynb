### Определяем город

```python
# загрузим данные с листа "города" справочника
cities_guide = pd.read_csv('https://docs.google.com/spreadsheets/d/' +
                   '1ODxYYhWq1OMntpGo9gD8jYKJQV6lGfw0fAtkyhCr1Q0' +
                   '/export?gid=0&format=csv')
```

Применим алгоритм нечеткого поиска - будем использовать метод process.extractOne с параметром частичного поиска - выбираем одно значение, которое больше всего подходит к искомому, возвращаем его и оценку результата, где 100 - значит 100% совпадение.

```python
# применим метод process, вернем наиболее совпадающее значение,
# отнесем в отдельные столбцы для дальнейшей работы с датафреймом
df[['city', 'city_score']] = df['Город'].apply(lambda x: process.extractOne(x, cities, scorer=fuzz.partial_ratio)) \
                                        .apply(pd.Series)
```

Видим, что есть подопечные из иностранных государств - отнесем их в категорию "Другая страна". Мне кажется важным отразить тот факт, что деятельность фонда знакома и людям, находящимся за рубежом. Также есть подопечные из других регионов РФ, но их адреса не найдены в справочнике ввиду отсутствия или неправильного написания.

```python
# создадим список иностранных государств, которые мы обнаружили в нашем датасете
countries = ['Казахстан', 'Кахахстан', 'Украина', 'Белорусия', 'Беларусь', 'Турция', 'Кипр', 'Эстония', 'Латвия', \
             'Таджикистан', 'Австрия', 'Великобритания', 'Великобретания', 'Канада', 'Абхазия', 'Молдавия', \
             'Германия', 'Киргизия']
```

```python
# напишем функцию, которая проверит совпадение значения в столбце с данными списка и вернет True/False
def check_country(text):
    for country in countries:
        if country in text:
            return True
    return False
```

```python
# применяем функцию к столбцу 'Город' и добавляем новый столбец 'other' - он будет техническим,
# в него будет записываться значение "Другая страна", если данные адреса содержатся в списке стран выше
df['other'] = df['Город'].apply(lambda x: 'Другая страна' if check_country(x) else '')
```

```python
# перезаписываем значения в столбце 'city'
df.loc[df['other'] == 'Другая страна', 'city'] = 'Другая страна'
```

```python
# пройдемся по строкам, где city_score < 90 и исправим, грубые ошибки (где строка отнесена к другому субъекту,
# и пустые строки, которые отнесены к г.Казань (т.к. "не указано")

df.loc[df['ФИО'] == 'User_01_71', 'city'] = 'Антрацит'
df.loc[df['ФИО'] == 'User_01_176', 'city'] = 'Зеленодольск'
df.loc[df['ФИО'] == 'User_01_623', 'city'] = 'Макеевка'
df.loc[df['ФИО'] == 'User_01_744', 'city'] = 'Тюмень'
df.loc[df['ФИО'] == 'User_01_73', 'city'] = 'Луганск'
df.loc[df['city'] == 'Сим', 'city'] = 'Симферополь'
df.loc[df['city'] == 'Арск', 'city'] = 'Самара'


city_unknown = ['User_01_293', 'User_01_1439', 'User_01_464', 'User_01_564', 'User_01_816', 'User_01_1099', \
                'User_01_1224', 'User_01_1334', 'User_01_1341', 'User_01_455', 'User_01_215', 'User_01_227', 'User_01_238']

df.loc[df['ФИО'].isin(city_unknown), 'city'] = 'Не указано'
```

Напишем функцию, которая будет присваивать значение из словаря, если найдет в подстроке соответствующее значение.

```python
def find_substring(x, dict):
  # начальное значение ключа - если подстрока не найдется, то значение вида сферы деятельности будет - Другая
    other_key = 'Другая'
    # пройдем по каждому элементу словаря, для каждого ключа проверим каждое значение в списке
    for key, values in dict.items():
        for value in values:
          # ищем любое совпадение - если совпадание найдено, присваиваем переменной соответственное значение ключа,
          # останавливаем перебор, переходим к следующей строке
            if pd.Series(x).astype(str).str.contains(value, regex=True, case=False).any():
                other_key = key
                break
    # функция возвращает значение переменной other_key - если совпадание нашлось, то это будет название категории,
    # если не нашлось - то Другая
    return other_key
```

```python
# применим функцию к столбцу "сфера деятельности" датафрейма
df['job'] = df['Сфера деятельности'].apply(lambda x: find_substring(x, busy_dict))
```

### Выделим стадию заболевания

Стадии заболевания обозначены как римские цифры, арабские или цифры в составе классификации TNM после буквы T.

```python
# создадим регулярное выражение для поиска римских цифр от I до IV (начинаем с 4, чтобы ее можно было найти,
# далее ищем 1, которая повторяется от 1 до 3 раз)
pattern_3 = re.compile(r'(IV|I{1,3})')

# Функция для замены римских цифр на арабские
def replace_roman(match):
    return str(roman.fromRoman(match.group(0)))
```

```python
pattern_4 = re.compile(r'(\d)\s*(?:стадия|ст)\b|\b(?:стадия|ст)\s*(\d)|\b(\d)(\D+|$)|(?:t|т|ршм)\s*(\d)', re.IGNORECASE)

# Извлекаем цифры из каждой строки столбца
df['grade'] = df['diagnose'].apply(lambda x: re.sub(pattern_3, replace_roman, x))\
.apply(lambda x: re.search(pattern_4, x).group(1) or re.search(pattern_4, x).group(2) or re.search(pattern_4, x) \
.group(3) or re.search(pattern_4, x).group(5) if re.search(pattern_4, x) else 'Не указана')
```

```python
# выведем срез таблицы после применения функции
df[['ФИО', 'Этап', 'stage']].head(15)
```

Мы просмотрели все строки, и по отдельным подопечным есть неточности. Далее исправим отдельные значения.

```python
# ID пациентов, по которым стадию лечения необходимо исправить на "химиотерапию"
chem = ['User_01_1418', 'User_01_984', 'User_01_946', 'User_01_607', 'User_01_1296']

# ID пациентов, по которым стадию лечения необходимо исправить на "начальный"
first = ['User_01_712', 'User_01_110', 'User_01_29']

# заменим значения этапа на корректные
df.loc[df['ФИО'] == 'User_01_1349', 'stage'] = 'Ремиссия'
df.loc[(df['ФИО'] == 'User_01_454') | (df['ФИО'] =='User_01_1395'), 'stage'] = 'Другая'
df.loc[df['ФИО'].isin(first), 'stage'] = 'Начальный'
df.loc[df['ФИО'].isin(chem), 'stage'] = 'Химиотерапия'
```

### Заполнение анкеты: за себя или за 3 лицо

```python
# создадим список слов, которые могут обозначать, что анкета заполнялась за 3 лицо - родственника или друга
who_list = ['мама', 'мамы', 'муж', 'супруг', 'доче', 'дочь', 'дочки', 'отец', 'отца', 'папы', 'папа', 'друг',
            'сын', 'тетя', 'тети', 'болеет', 'сын', 'внук', 'сестр', 'брат', 'тёти']
```

Проверим вхождение подстроки из списка в строку с помощью функции pandas contains(). В функции каждую строку предварительно переводим в Series, т.к. contains() работает с типом объектов Series.

```python
# напишем функцию, которая проверяет вхождение слов из списка who_list в строку, переведенную в Series
df['applicant'] = df['diagnose'].apply(lambda x: 'За третье лицо' if \
                                       any(val in x or pd.Series(x).str.contains(val, regex=True, case=False)\
                                       .any() for val in who_list) else 'За себя')
```

```python
# отобразим срез таблицы
df[['Диагноз', 'applicant']].head()
```

```python
# создадим сводную таблицу и нарисуем график, задав его размеры, заголовок и подписи оси
appl_plot = df.pivot_table(index='applicant', values='ФИО', aggfunc='count')\
              .reset_index().sort_values(by='ФИО', ascending=False)
plt.figure(figsize=(7, 5))
plt.title('заявитель')
plt.xlabel('количество анкет')
sns.barplot(appl_plot, x='ФИО', y='applicant', orient='h', color='lightblue');
```
